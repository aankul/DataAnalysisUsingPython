Question 1 (a) : Create a list of numbers which are multiple of 3 and 7 between 1 to 500?

```python
multiList = []
for i in range(1,500):
    if(i%21==0):
        multiList.append(i)
print ('The list of numbers which are multiple of 3 & 7 are : \n{multiList}'.format(multiList = multiList))
```

Question 1 (b) : For the list, find what is the mean. Print what is 25th quartile, 50th quartile and 75th quartile?

```python
l = len(multiList)
print('The mean is : {m}'.format(m=sum(multiList)/l))
print('25th quartile :{q25}'.format(q25=multiList[l//4]))
print('50th quartile :{q25}'.format(q25=multiList[l//2]))
print('75th quartile :{q25}'.format(q25=multiList[3*l//4]))
```

Question 2 (a) : Read the file ‘ques2.txt’. Print the number of words.

```python
myfile = open('ques2.txt')
wordList = myfile.readline()
count = 0;
for word in wordList.split() :
    count = count + 1
print('The number of word is : {count}'.format(count=count))
```



Question 2 (b) : Find out what is the average, median and mode for word length (character length).

```python
lengthList = []
for word in wordList.split() :
    lengthList.append(len(word))
lengthList.sort()
m=sum(lengthList)/len(lengthList)
print('The mean of the list is {m}'.format(m=m))
print('The median of the list is {median}'.format(median=lengthList[len(lengthList)//2]))
print('The mode of the list is {mode}'.format(mode = max(set(lengthList), key=lengthList.count)))
```
